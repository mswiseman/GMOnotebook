# Run script to produce fluorescent protein signal plots over *all tissue combined* in **leaf** explants

Leaf explants are automatically excluded from the standard workflow because we do not currently have a segmentation model trained to segment leaf explants from tissues derived thereof.<br>

Before running this script, make sure that the randomization datasheet and missing explant sheet are provied and formatted appropriately. Note there should be a single line per sample, including in cases where an image is re-taken. Also, note this workflow will only analyze explants labeled as tissue type "L" in the randomization datasheet.

The output plots will be saved into the directory "4B. GMOlabeler statistic plots" linked on the desktop.

This script contains a large number of debugging statements that the user does not need to worry about. Sorry for the mess! These will be removed later, once the code is developed beyond a beta stage. The only output below that matters to the user is an error message (if there is a problem) or the path to the raw data spreadsheet. These will appear at the *very bottom*.

The `any_missing` variable should be set to 1 if any explants are missing, or 0 if not. In the case of missing explants, the missing explant `csv` must also be provided.

Set variables below.

```bash
data_folder=Elements_10//Transformation/GRF_Poplar/GRF1_Phase_2/wk3/
randomization_datasheet=/run/media/labgroup/Elements_10//Transformation/GRF_Poplar/GRF1_Phase_2/GRF_Hyperspectral_Labels_P2.xlsx
any_missing=1
missing_explants=/run/media/labgroup/Elements_10//Transformation/GRF_Poplar/GRF1_Phase_2/wk3/missing_explants.csv
grid=20
height=5
width=9
```

Run this script using variables provided above. There is no need to modify the below script call.

```bash
cd /home/labgroup/code/GMOnotebook/Extras
```

```bash
conda activate gmolabeler
Rscript Leaf_explant_plots.R \
-d "${data_folder}/" \
-r "$randomization_datasheet" \
-m $any_missing \
-M $missing_explants \
-g $grid \
--sort 1 \
--height $height \
--width $width
conda deactivate
```
