# Determine grid borders for hyperspectral data

We must determine the grid borders of a hyperspectral image (or equivalently, an RGB image aligned to a hyperspectral image) similarly to how we did previously for unaligned RGB data in notebook `1_Find_pre_aligned_rgb_grid_borders.ipynb`.

You may choose to do this either using a jpg of a channel extracted from a hyperspectral image, or using an RGB image aligned with a hyperspectral data. The latter may show edges somewhat more clearly, and is thus the recommended approach.

In the screenshots below, we will again use GNU Image Manipulation Program (GIMP) to find the grid borders. This time, the image *does not* need to be rotated or re-scaled by the user.

Let's open our aligned version of the grid standard, which will have the suffix `aligned.jpg` after running notebook `3A-Compute_homography_scikit.ipynb`.<br>

***Top:*** When we point the cursor at the middle of the top border, we can see on the status bar (bottom) that our cursor is at x,y position 788,126. For the top and bottom borders, we are interested in the y position. Thus, the top grid border position is **126**.<br>


***Bottom:*** With cursor at the middle of the bottom border, we can see that the position is **1278**.<br>

***Left:*** With the cursor at the middle of the left border, we can now see the cursor is at x,y position 320,708. For left and right borders, we're interested in the x position. Thus, the left grid border position is **320**.<br>

***Right:*** With cursor at the middle of the right border, we can see that the position is **1252**.<br>
